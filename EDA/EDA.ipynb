# COPD in Nepal
### Factors that lead to COPD
1. Biomass Fuel Exposure (Indoor Air Pollution)
2. Smoking Tobacco
3. Outdoor Air Pollution in Urban Areas
4. Occupational Exposure
5. Genetic susceptibility
6. Respiratory Infection
7. Low Awareness and Health Access
8. Climate & Environmental factors

# Sign & Symptoms of COPD
1. Chronic Cough
2. Shortness of Breath (Dyspnea)
3. Excessive mucus production (Sputum) (Phlegm)
4. Wheezing
5. Fatigue & Low energy
6. Tightness in Chest
7. Frequent Respiratory Infections
8. Cynosis (Blusih Lips or Fingernail beds)
9. Unintended Weight Loss
10. Chronic Bronchitis & Emphysema Symptoms
11. Frequent Flare-ups (Exacerbations)

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = r'/Users/roshanthapa/Desktop/Omdena-Capacity_Building/copd/COPD_Prediction/Data/synthetic_COPD_data.csv'
df = pd.read_csv(path)
df.head()
```

```python
# Set up a large plot for better visibility
plt.figure(figsize=(14, 8))
```

```python
# Get information about my data
df.info()
```

# Univariate Analysis

```python
# What is the age distribution in my data?
sns.histplot(df['Age'], kde=True, color='blue')
plt.title('Age distribution')
```

```python
# BMI Distribution
sns.histplot(df['BMI'], kde=True, color='green')
plt.title('BMI Distribution')
```

```python
# Univariate analysis for the categorical features
# gender distribution
sns.countplot(x='Gender', data=df, palette='viridis')
plt.title('Gender Distribution')
```

```python
# Smoking Status
sns.countplot(x='Smoking_Status', data=df, palette='Set1')
plt.title('Smoking Status Distribution')
```

# Bivariate Analysis

```python
plt.figure(figsize=(14,10))
```

```python
# 1. Age vs COPD Diagnosis
sns.boxplot(x='COPD_Diagnosis', y='Age', data=df, palette='cool')
plt.title('Age vs COPD Diagnosis')
plt.show()
```

```python
# 2. COPD Diagnosis vs BMI

sns.boxplot(x='COPD_Diagnosis', y='BMI', data=df, palette='coolwarm')
plt.title('COPD Diagnosis vs BMI')
plt.show()
```

```python
sns.countplot(x='COPD_Diagnosis', data=df, palette='viridis')
plt.title('Count plot of COPD Diagnosis')
plt.show()
```

```python
# Correlation Matrix
data_corr = df[['Age', 'Biomass_Fuel_Exposure', 'Occupational_Exposure','Family_History_COPD','BMI','Air_Pollution_Level','Respiratory_Infections_Childhood','COPD_Diagnosis']]
corr = data_corr.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation matrix')
plt.show()
```

```python
# pair plot
sns.pairplot(df, hue='COPD_Diagnosis',diag_kind='kde', palette='Set1')
plt.title('Pair plot')
plt.show()
```
